```python
import seaborn as sb
```

```python
%matplotlib inline
```

```python
tips=sb.load_dataset('tips')
tips
```

```python
sb.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
 sb.jointplot(x='total_bill',y='tip',data=tips,kind="hex")
```

```python
sb.pairplot(tips,hue='sex')
```

```python
sb.rugplot(tips['total_bill'])
```

```python
sb.kdeplot(tips["total_bill"]) 
```

```python
sb.barplot(x='sex',y='total_bill',data=tips)
```

```python
sb.countplot(x='sex',data=tips,hue="smoker")
```

```python
sb.boxplot(x='day',y='total_bill',data=tips)
```

```python
sb.stripplot(x='day',y='total_bill',data=tips)
```

```python
sb.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sb.factorplot(x="day",y="total_bill",data=tips,kind="bar")
```

```python
iris= sb.load_dataset("iris")
iris
```

```python
sb.pairplot(iris)
```

```python
g=sb.PairGrid(iris)
g.map_diag(sb.distplot)
g.map_lower(sb.kdeplot)
g.map_upper(sb.kdeplot)
```
